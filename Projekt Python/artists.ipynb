```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
torrenty = pd.read_csv('torrenty_202106201241.csv')
```

```python
torrenty.head()
```

```python
torrenty.info()
```

```python
torrenty.describe()
```

```python
torrenty['tag'] = torrenty['tag'].astype(str)
```

```python
torrenty_title_merged = torrenty.groupby(['artist', 'groupName']).agg({
    'totalSnatched': 'first', 
    'groupYear': 'first',
    'releaseType': 'first',
    'tag': (';').join}
)
torrenty_title_merged.head(2)
```

```python
tag_dummies = pd.get_dummies(torrenty['tag'])
tag_dummies.head(1)
```

```python
torrenty_tag_split = torrenty.drop(['tag'], axis=1).join(tag_dummies)
```

```python
torrenty_title_merged_1 = torrenty_tag_split.iloc[:,:6].groupby(['artist', 'groupName']).agg({
    'totalSnatched': 'first', 
    'groupYear': 'first',
    'releaseType': 'first'
})
torrenty_title_merged_2 = torrenty_tag_split.iloc[:,:].groupby(['artist', 'groupName']).agg({
    'max'
}).iloc[:,4:]
```

```python
torrenty_title_merged = torrenty_title_merged_1.join(torrenty_title_merged_2)
torrenty_title_merged.head(2)
```

```python
torrenty_title_merged.info()
```

```python
torrenty_title_merged = torrenty_title_merged.reset_index()
```

Number of unique artists

```python
torrenty_title_merged['artist'].unique().size
```

Number of titles for an artist

```python
torrenty_title_merged.groupby('artist').agg({'groupName': 'count'}).describe()
```

Span of years for an artist

```python
torrenty_title_merged.groupby('artist').agg({'groupYear': ['max', 'min']}).reset_index().apply(lambda x: int(x[1])-int(x[2]), axis=1).describe()
```
